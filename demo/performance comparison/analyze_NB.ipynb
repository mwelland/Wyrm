```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#import plotly.express as px


df_mg = pd.read_csv("walltime_GM_3D.txt",sep=' ',header=0)
df_direct = pd.read_csv("walltime_direct_3D.txt",sep=' ',header=0)

plt.plot(df_direct['ndf'],df_direct['walltime']/60,label='Direct')
plt.plot(df_mg['ndf'],df_mg['walltime']/60,label='MG')
plt.legend()
plt.xlabel("Degrees of Freedom")
plt.ylabel("wall time (min)")
plt.show()
```
